```python
import pandas as pd
import numpy as np

pd.set_option('display.max_rows', 100)

# for plotting, not truly used in this notebook
import matplotlib.pyplot as plt
import seaborn as sns
```

Based on the survey statistics, generate barchar


```python
path = "../data/"
file1 = "users_stats.csv"
file2 = "movies_stats.csv"
file3 = "list of movies - dataset.csv"

users_stats = pd.read_csv(path+file1)
movies_stats = pd.read_csv(path+file2)
movies_imdb = pd.read_csv(path+file3)
```

```python
users_count = users_stats.sort_values(by=["count"], ascending = False)[["userId","count"]]
```

```python
plt.figure(figsize=(10,10))
sns.set(style="whitegrid")
ax = sns.barplot(x="count", y="userId", data=users_count, palette = "Blues_d")
ax.set(ylabel="",xlabel="")
sns.despine(left=True, bottom=True, right=True)
```

```python
users_rateing_avg = users_stats.sort_values(by=["mean"], ascending = False)[["userId","mean"]]
```

```python
plt.figure(figsize=(10,10))
sns.set(style="whitegrid")
ax = sns.barplot(x="mean", y="userId", data=users_rateing_avg, palette = "ch:2.5,-.2,dark=.3")
ax.set(ylabel="",xlabel="")
sns.despine(left=True, bottom=True, right=True)
```

```python
plt.figure(figsize=(10,30))
ax = sns.barplot(x="count", y="Unnamed: 0", data=movies_stats.sort_values(by="count", ascending=False).head(98), palette = "ch:3.5,-.2,dark=.3")
ax.set(ylabel="",xlabel="")
sns.despine(left=True, bottom=True, right=True)
```

```python
plt.figure(figsize=(10,5))
sns.set(style="whitegrid")
ax = sns.barplot(x="reviews",
                 y="title_survey",
                 data=movies_imdb.sort_values(by="reviews", ascending=False).head(10),
                 palette = "ch:1.5,-.2,dark=.3")
ax.set(ylabel="",xlabel="")
sns.despine(left=True, bottom=True, right=True)
```
